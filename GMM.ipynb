```python
from sklearn.cluster import KMeans
```

```python
for k in range (1, 81,10):
 
	# Create a kmeans model on our data, using k clusters.  random_state helps ensure that the algorithm returns the same results each time.
	kmeans_model = KMeans(n_clusters=k, random_state=1).fit(scaledData)
	
	# These are our fitted labels for clusters -- the first cluster has label 0, and the second has label 1.
	labels = kmeans_model.labels_
 
	# Sum of distances of samples to their closest cluster center
	interia = kmeans_model.inertia_
	print ("k:",k, " cost:", interia)
```

```python
from sklearn.cluster import MiniBatchKMeans
import time
```

```python
# Compute clustering with MiniBatchKMeans
for k in range (1,142,10):
    mbk = MiniBatchKMeans(init='k-means++', n_clusters=k, batch_size=100,random_state=1)
    t0 = time.time()
    model = mbk.fit(scaledData)
    t_mini_batch = time.time() - t0
    interia = model.inertia_
    print ("k:",k, " cost:", interia)
    print(t_mini_batch)
```

```python
# Compute clustering with MiniBatchKMeans
for k in range (1,150,1):
    mbk = MiniBatchKMeans(init='k-means++', n_clusters=k, batch_size=100,random_state=1)
    t0 = time.time()
    model = mbk.fit(scaledData)
    t_mini_batch = time.time() - t0
    interia = model.inertia_
    print ("k:",k, " cost:", interia)
    print(t_mini_batch)
```
